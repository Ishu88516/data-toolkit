#Theory questions

1. What is NumPy and why is it widely used?

NumPy is a Python library for numerical computing. It provides fast, memory-efficient arrays and mathematical functions.

2. How does broadcasting work in NumPy?

Broadcasting allows NumPy to perform operations on arrays of different shapes by expanding smaller arrays automatically.

3. What is a Pandas DataFrame?

A DataFrame is a 2D labeled data structure (rows & columns) similar to a table or spreadsheet.

4. Purpose of groupby() in Pandas

It groups data and applies aggregation functions like sum(), mean(), or count().

5. Why is Seaborn preferred for statistical visualization?

It provides attractive default styles and easy statistical plotting.

6. NumPy arrays vs Python lists
NumPy Array	Python List
Faster	Slower
Fixed data type	Mixed types
Vectorized ops	Loop-based
7. What is a heatmap?

A heatmap visualizes data values using colors, commonly used for correlation matrices.

8. What is a vectorized operation?

Operations applied to entire arrays without explicit loops.

9. Matplotlib vs Plotly

Matplotlib: Static, customizable

Plotly: Interactive, web-based

10. Hierarchical indexing in Pandas

Allows multiple index levels for complex data representation.

11. Role of Seaborn’s pairplot()

Shows pairwise relationships between variables.

12. Purpose of describe()

Provides summary statistics (mean, min, max, quartiles).

13. Why handle missing data?

Missing data can cause incorrect analysis or errors.

14. Benefits of Plotly

Interactive charts, zooming, tooltips, dashboard-friendly.

15. How NumPy handles multidimensional arrays

Using ndarray objects with ndim and shape attributes.

16. Role of Bokeh

Creates interactive, browser-based visualizations.

17. apply() vs map()

map() → Series only

apply() → Series or DataFrame

18. Advanced NumPy features

Broadcasting, linear algebra, FFT, random sampling, masked arrays.

19. Pandas and time series

Built-in datetime support, resampling, rolling windows.

20. Role of pivot tables

Summarizes and aggregates large datasets.

21. Why NumPy slicing is faster

It uses views instead of copying data.

22. Common Seaborn use cases

Statistical plots, distribution analysis, regression plots, correlation heatmaps.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = sns.load_dataset("iris")
# Drop non-numeric columns like 'species' before calculating correlation
corr = df.drop('species', axis=1).corr()

sns.heatmap(corr, annot=True)
plt.show()
```
